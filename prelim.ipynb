```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
def slope(betaL, kappa, kappa_c):

    tau = np.sqrt(1 - kappa**2)
    tau_c = np.sqrt(1 - kappa_c**2)

    gamma = (kappa**2 * tau_c) / (2 * tau)
    rho = (kappa_c * (1 + tau**2)) / (2 *tau)
    
    num = 8 * gamma**2 * (np.sin(betaL) + rho) * np.cos(betaL) * (gamma**4 - (np.sin(betaL) + rho)**4)
    den = ((np.sin(betaL) + rho)**2 + gamma**2)**4

    return num / den
```

```python
SAMPLES = 1000
_x = np.linspace(0.0001, 0.9999, SAMPLES)
x, y = np.meshgrid(_x, _x)
```

```python
BETA_L_INIT = -.794 # this is arbitrary, just checked which betaL has the highest slope by eye lol
__x = np.linspace(-np.pi, np.pi, SAMPLES)
kappa = np.random.random()
kappa_c = np.random.random()

print(kappa, kappa_c)

fig, ax = plt.subplots()
plot = ax.contourf(x, y, slope(BETA_L_INIT, x, y), levels=50)
fig.colorbar(plot)

print(slope(BETA_L_INIT, .144, .720))
ax.set_xlabel(r"$\kappa$");
ax.set_ylabel(r"$\kappa_c$");
ax.set_xlim(0, 0.1);
ax.set_ylim(0.710, 0.716);
# ax.plot(__x, slope(__x, kappa, kappa_c))
```
